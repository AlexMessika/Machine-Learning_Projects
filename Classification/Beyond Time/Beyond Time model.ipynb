```
head = ["question1","question2","question3","question4","question5","question6","question7","question8","question9","question10","question11","question12",]
```

```
def convert_void(x):

  if x not in [0, 1, 2, 3, 4]:
    return 2.
  return float(x)

def convert_conseil(x):
  if x == "Formation": return float(1)
  if x == "Primes": return float(2)
  if x == "Package": return float(3)
  if x == "Espace détente": return float(4)
  if x == "Payer Uber - Télé travail": return float(5)
  if x == "Sanction": return float(6)
  if x == "Temps de pause": return float(7)
  if x == "Isolement": return float(8)
  if x == "Rien à dire": return float(9)
  if x == "Entretien": return float(10)
  if x == "Horaires atypiques": return float(11)
  if x == "Team building": return float(12)
  if x == "Campagne de communication": return float(13)
```

```
X = pd.DataFrame(columns = head)
for i in range(12):
  X[head[i]] = df_RH.iloc[:,i].apply(convert_void)
```

```
y = df_RH["Conseil"]
y = pd.DataFrame(columns=["Conseil"])
y["Conseil"] = df_RH["Conseil"].apply(convert_conseil)
```

```
from sklearn.model_selection import train_test_split
    

X_train, X_test, y_train, y_test = None, None, None, None
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=.2, random_state=0)
y_train = np.ravel(y_train)
y_test = np.ravel(y_test)
print( X_train.shape)
print( X_test.shape)
print( y_train.shape)
print( y_test.shape)
```

```
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=10)
neigh.fit(X_train, y_train)
```

```
neigh.score(X_test,y_test)
```

```
accuracy = []
k = []
for i in range (1, 50):
  knn = KNeighborsClassifier(n_neighbors = i)
  knn.fit(X_train, y_train)
  accuracy.append(knn.score(X_test, y_test))
  k.append(i)
```

```
plt.figure()
plt.plot(k, accuracy)
```
